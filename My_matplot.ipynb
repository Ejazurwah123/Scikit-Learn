Matplot WorkFlow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# set up plot
fig, ax = plt.subplots(figsize=(10,10))

#plot the data
ax.plot(x,y)

# customise the plot
ax.set(title='Simple Plot', xlabel='X-axes',ylabel='y axes')

# display
plt.show()
```

```python
import numpy as np
x = np.linspace(0,10,100)
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(x, np.sin(x));
```

```python
# making graph from dictionary

my_dict = {
    'Urwah' : 50,
    'Marwa' : 60,
    'Haura' : 70
}
fig, ax = plt.subplots()
ax.bar(my_dict.keys(),my_dict.values());
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2,figsize=(10,5))
ax1.scatter(np.linspace(20,300,20),np.linspace(20,300,20));
```
